# 量子力学中的 Schrieffer-Wolff 变换 (SWT)

### 1. 核心概念
**Schrieffer-Wolff 变换 (SWT)** 是一种二阶微扰理论的算符化方法。它的核心目的是：通过一个**幺正变换（Unitary Transformation）**，将哈密顿量中高低能级子空间之间的微弱耦合“旋转掉”，从而得到一个在低能子空间内独立的**有效哈密顿量（Effective Hamiltonian）**。

* **直观理解**：就像在处理一个带有细微倾斜的坐标系，通过旋转坐标轴，让原本混合在一起的变量变得相互独立。
* **应用场景**：超导量子比特（Transmon）的色散读取、自旋-轨道耦合、近藤模型（Kondo Model）等。

---

### 2. 数学推导简述
假设系统的哈密顿量为：
$$H = H_0 + V$$
其中 $H_0$ 是对角项（无扰动），$V$ 是非对角扰动项（耦合）。

我们寻找一个幺正算符 $U = e^S$（$S$ 是反埃尔米特算符，$S^\dagger = -S$），使得变换后的哈密顿量 $H_{eff}$ 尽可能对角化：
$$H_{eff} = e^S H e^{-S}$$

利用 **BCH 展开**：
$$H_{eff} = H_0 + V + [S, H_0] + [S, V] + \frac{1}{2}[S, [S, H_0]] + \dots$$

为了消除一阶非对角项 $V$，我们令：
$$[S, H_0] = -V$$

代入展开式，一阶项抵消，保留到二阶的有效哈密顿量为：
$$H_{eff} \approx H_0 + \frac{1}{2}[S, V]$$

---

### 4. 结论总结
1. **物理本质不变**：`H` 和 `H_trans` 的本征值完全一样。证明了 SWT 只是一个基底变换，没有改变物理系统。
2. **解耦作用**：变换后的矩阵非对角项趋于 0。
3. **能量修正**：SWT 二阶近似通过计算 $\frac{g^2}{\Delta}$ 捕捉到了能级的位移。

```python
import numpy as np
from scipy.linalg import expm

# ==========================================
# 1. 系统参数定义
# ==========================================
Delta = 10.0  # Z 方向主能级差 (Energy Gap)
g = 0.5       # X 方向的微弱耦合 (Perturbation)

# 泡利矩阵
sz = np.array([[1, 0], [0, -1]])
sx = np.array([[0, 1], [1, 0]])

# 原始哈密顿量 H = H0 + V
H0 = (Delta / 2) * sz
V = g * sx
H = H0 + V

# ==========================================
# 2. Schrieffer-Wolff 变换实现
# ==========================================
# 根据 [S, H0] = -V 求解生成元 S
# 对于单管比特系统，S = (g/Delta) * [0, 1; -1, 0]
S = (g / Delta) * np.array([[0, 1], [-1, 0]])

# 计算幺正矩阵 U = exp(S)
U = expm(S)

# A. 精确幺正变换 (Base Change)
# H_transformed = U * H * U_dagger
H_trans = U @ H @ expm(-S)

# B. SWT 二阶近似公式
# H_eff = H0 + 0.5 * [S, V]
commutator_SV = S @ V - V @ S
H_eff_approx = H0 + 0.5 * commutator_SV

# ==========================================
# 3. 结果验证与打印
# ==========================================
print("--- 矩阵展示 ---")
print(f"原始 H (包含非对角项 g):\n{H}\n")
print(f"变换后的 H (非对角项被消除):\n{np.round(H_trans, 5)}\n")

# 获取能级（本征值）
eig_orig = np.sort(np.linalg.eigvals(H).real)
eig_trans = np.sort(np.linalg.eigvals(H_trans).real)
eig_approx = np.sort(np.diag(H_eff_approx).real)

print("--- 等效性证明 (能量谱对比) ---")
print(f"原始矩阵精确能级:   {eig_orig}")
print(f"幺正变换后精确能级: {eig_trans}")
print(f"SWT 二阶近似能级:   {eig_approx}")

# 计算误差
error = np.abs(eig_orig - eig_approx)
print(f"\n微扰近似误差: {error}")
```
