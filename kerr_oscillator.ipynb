```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt

# ====================== 1. 参数 ======================
N = 15          # 希尔伯特空间维度
omega = 1.0     # 线性频率
K = -0.1        # Kerr 系数（Transmon 是负的）

# ====================== 2. 算符 ======================
a = qt.destroy(N)
ad = a.dag()
n = ad * a

# ====================== 3. Kerr 哈密顿量 ======================
H = omega * (ad * a) + 0.5 * K * (ad * ad * a * a)
```

```python


# ====================== 4. 时间演化 ======================
psi0 = qt.coherent(N, 2.0)          # 初始相干态
tlist = np.linspace(0, 20, 100)
result = qt.mesolve(H, psi0, tlist) # 演化

# ====================== 5. 画 Wigner 函数（修复版） ======================
final_state = result.states[-1]

# 计算 Wigner 函数数据
xvec = np.linspace(-4, 4, 200)
W = qt.wigner(final_state, xvec, xvec)

# 画热图
plt.figure(figsize=(6,5))
plt.contourf(xvec, xvec, W, 100, cmap='RdBu')
plt.colorbar()
plt.title("Kerr Oscillator Wigner Function")
plt.axis("equal")
plt.show()
```

```python
# 初始态：相干态（经典类似“正弦波”）
psi0 = qt.coherent(N, 2.0)

# 时间点
tlist = np.linspace(0, 20, 100)

# 演化
result = qt.mesolve(H, psi0, tlist)

# 画 Wigner 函数（看 Kerr 非线性导致的量子态畸变）
qt.wigner_cmap(result.states[-1])
qt.wigner(result.states[-1])
plt.title('Wigner function of Kerr oscillator')
plt.show()
```
